```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(10)

df=pd.DataFrame({
    "Name":['student_'+str(i) for i in range(1,101)],
    "Class":np.random.choice(['10A','10B','10C'],100),
    "Gender":np.random.choice(['Male','Female'],100),
    "Maths":np.random.randint(35,100,100),
    "Science":np.random.randint(35,100,100),
    "English":np.random.randint(35,100,100),
    "Attendence_%":np.random.randint(60,100,100)

})

df.loc[[5,18,33,67,89],"Science"]=np.nan

print(df.head(10))
```

```python
df
```

```python
df=df.assign(Total_marks=df["Maths"]+df["Science"]+df["English"])
```

```python
df
```

```python
df=df.assign(Percentage=(df["Total_marks"]/300)*100).round(2)
```

```python
df
```

```python
df["Result"]=df["Percentage"].apply(lambda x:"pass" if x>=40 else "fail")
```

```python
df
```

```python
avg_science_marks=df["Science"].mean()
df=df.fillna(avg_science_marks)
```

```python
df.head(10)
```

```python
df.groupby("Class")["Percentage"].mean().idxmax()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
avg_maths=df.groupby("Class")["Maths"].mean()
```

```python
plt.bar(avg_maths.index,avg_maths.values)
plt.title("Class vs Average Maths Marks")
plt.xlabel("Class")
plt.ylabel("Avg maths marks")
```

```python
df.boxplot(column='English',by='Gender')
plt.title("Gender vs English Marks")
plt.suptitle("")
plt.xlabel("Gender")
plt.ylabel("English Marks")
plt.show()
```

```python
result_count=df["Result"].value_counts()
plt.pie(result_count,autopct="%1.1f%%",labels=result_count.index)
plt.legend()
plt.title("Pass Vs Fail")
```

```python
percentage_data=df["Percentage"]
plt.hist(percentage_data,edgecolor="black",bins=20)
```

```python
Attendence=df["Attendence_%"]
percentage=df["Percentage"]
plt.scatter(Attendence ,percentage)
plt.title("Attendence_% Vs Percentage")
plt.xlabel("Attandence_%")
plt.ylabel("Percentage")
```
