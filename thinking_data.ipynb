```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)

    return data
```

```python
data = load_data("store_data.json")
print(data)
```

```python
def clean_data(data):
    cleaned_data=[]
    unique_data=set()
    
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    for user in data:
        raw_rating=str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating=text_to_num[raw_rating]

        user["rating"]=raw_rating

        raw_age=user.get("age")
        if(raw_age==None):
            user["age"]=None

        if(user["name"].strip() in unique_data):
            continue

        unique_data.add(user["name"])
        cleaned_data.append(user)

    return cleaned_data
```

```python
clean_data(data)
```

```python
def get_insights(data):
   tot_rating=0
   for user in data:
       tot_rating+=int(user["rating"])

   print(f"avg of rating {tot_rating/len(data)}")

   poor_rating=0
   for user in data:
       if(int(user["rating"]) < 3):
           poor_rating+=1
   print(f" % of poor rating {(poor_rating/len(data))*100}")
```

```python
get_insights(data)
```

```python
def get_recommendations(data):
    recommendations =[]
    for user in data:
        curr_rec= {}
        curr_rec["name"]=user["name"]
        if (int(user["rating"]) >=4):
            curr_rec["brand"]="Apple"
        else:
            curr_rec["brand"]="Samsung"
        recommendations.append(curr_rec)
    return recommendations
            
            
```

```python
get_recommendations(data)
```
