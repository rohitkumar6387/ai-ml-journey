```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame({
    'Date':['2026--05-01','2026--05-01','2026--05-02','2026--05-02','2026--05-03','2026--05-03','2026--05-03'],
    'Item':['Maggi','Biscuit','Maggi','Chips','Biscuit','Chips','Maggi'],
    'Qty_sold':[10,20,15,25,30,10,12],
    'Cost_price':[10,5,10,8,5,8,10],
    'Selling_Price':[12,7,12,10,7,10,12]
})
df['Date']=pd.to_datetime(df['Date'])
```

```python
df
```

```python
df=df.assign(Total_cost=df['Qty_sold']*df['Cost_price'])
df
```

```python
df=df.assign(Total_revenue=df['Qty_sold']*df['Selling_Price'])
df
```

```python
df=df.assign(Profit=df['Total_revenue']-df['Total_cost'])
df
```

```python
df2=df.groupby('Date')['Profit'].sum()
df2
```

```python
print("Best_day",df2.idxmax(),"with",df2.max(),"Rs.")
```

```python
df3=df.groupby('Item')['Qty_sold'].sum()
print("Best selling item-",df3.idxmax(),"-",df3.max(),"Unit")
```

```python
df4=df.groupby('Item')['Profit'].sum()
print("Most profitable item:",df4.idxmax(),"-",df4.max(),"Rs total profit")
```

```python
df['Profit_margin']=(df['Profit']/df['Total_revenue'])*100 
df
```

```python
df.to_csv("Kirana_sales_Report.csv",index=False)
```
