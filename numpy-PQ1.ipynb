```python
import numpy as np
```

```python
sales=np.array([120,150,90,200,180,300,250])
expenses=np.array([50,60,40,80,70,120,100])
```

```python
sum_s=np.sum(sales)
sum_exp=np.sum(expenses)
```

```python
profit=sum_s-sum_exp
print("Profit=",profit)
```

```python
max_sale=np.max(sales)
day=np.argmax(sales)
```

```python
print("Max_sale =",max_sale,", Day =",day)
```

```python
Daily_profit=sales-expenses
Daily_profit
avg_daily=np.mean(Daily_profit)
print("Average Daily profit",avg_daily)
```

```python
count=0
b=Daily_profit>avg_daily
a=0
while a<len(b):
    if b[a]==True:
        count+=1
    a+=1
print(count,"days,avg_profit>daily_profit")
    
```

```python
weekend_profit=np.sum(Daily_profit[5:])
per_weekend_profit=((weekend_profit/np.sum(Daily_profit))*100)
print("Percentage of weekend profit",per_weekend_profit,"%")
```
