```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
scores=np.random.normal(70,10,100)
plt.hist(scores,color="skyblue",edgecolor="black")
plt.axvline(
    x=70,
    label="mean value(70)",
    color="red",
    linewidth="2",
    linestyle="-."
)
        
plt.xlabel("scores")
plt.ylabel("number of students")
```

```python
legit_transaction=[
    2.99,3.50,3.99,4.23,5.50,5.99,6.99,8.50,9.43,10.67,11.99,12.89,13.50,15.00,18.99,
    20.00,23.33,25.50,27.89,30.00,35.67,38.99,45.00,50.00,53.33,59.88,66.00,75.87,80.90,
    100.10,120.00,135.54,150.00,250.50,453.34,548.98,673.56,898.00,999.99
]

fraud_transaction=[
    50,100,150,300,500,750,1000,1000,1200,1500,1500,2000,2200,2500,3000,3000
]
plt.hist(fraud_transaction,color="red",bins=20,label="Fraud",edgecolor="black",alpha=0.5)
plt.hist(legit_transaction,color="green",bins=20,label="Legit",edgecolor="black",alpha=0.5)
plt.title("Transaction Amount Distribution-Legit Vs Fraud")
plt.xlabel("Amount (in $)")
plt.ylabel("Frequency")
plt.legend()
```
