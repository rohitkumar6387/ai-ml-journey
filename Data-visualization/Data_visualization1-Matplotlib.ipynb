```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.plot(x,y)
```

```python
oscar_movies=[
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years=[2000,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232]
plt.plot(years,oscar_revenue)
plt.title("oscar movies revenue in each year")
plt.xlabel("Years")
plt.ylabel("Revenue(in SM)")
```

```python
non_oscar_movies=["slumdog Millonaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue=[378,2788,829,185,275]
plt.plot(years,oscar_revenue,'o--r',label="Oscar movies")
plt.plot(years,non_oscar_revenue,color="g",linestyle="--",label="Non_Osacar movies")
plt.title("oscar movies vs non_oscar revenue")
plt.xlabel("Years")
plt.ylabel("Revenue(in SM)")
plt.legend()
```
