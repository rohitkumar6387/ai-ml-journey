```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
days=["Mon","Tue","Wed","Thu","Fri"]

cities=["New York","London","Delhi","Tokyo"]

temperature=[
    [22,23,21,24,25],
    [18,19,17,20,21],
    [30,32,31,33,34],
    [25,26,24,27,28]
]
```

```python
fig,ax=plt.subplots(2,2,figsize=(7,7))

ax[0][0].plot(days,temperature[0])
ax[0][0].set_ylabel("temperature")
ax[0][0].set_xlabel("days")
ax[0][0].set_title("New york")

ax[0][1].plot(days,temperature[1])
ax[0][1].set_ylabel("temperature")
ax[0][1].set_xlabel("days")
ax[0][1].set_title("London")

ax[1][0].plot(days,temperature[2])
ax[1][0].set_ylabel("temperature")
ax[1][0].set_xlabel("days")
ax[1][0].set_title("Delhi")

ax[1][1].plot(days,temperature[3])
ax[1][1].set_ylabel("temperature")
ax[1][1].set_xlabel("days")
ax[1][1].set_title("Tokyo")

fig.suptitle("Days Vs Temperature (According to cities")
fig.tight_layout()
```

```python
#or 

fig,ax=plt.subplots(2,2,figsize=(7,7))

city_no=0
for i in range(2):
    for j in range(2):
        ax[i][j].plot(days,temperature[city_no])
        ax[i][j].set_title(cities[city_no])
        ax[i][j].set_xlabel("Days")
        ax[i][j].set_ylabel("Temperature")
        ax[i][j].grid(True)
        city_no+=1

fig.suptitle("Days vs Temperature(According to cities)")
fig.tight_layout()
```
