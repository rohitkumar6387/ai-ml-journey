```python
import matplotlib.pyplot as plt+
```

```python
oscar_movies=[
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232]
plt.bar(years,oscar_revenue)
plt.title("oscar movies revenue in each year")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")

for i in range(len(years)):
    plt.text(years[i],oscar_revenue[i]+10,str(oscar_revenue[i]),ha="center")
plt.ylim(0,max(oscar_revenue)+100)
```

```python
import numpy as np
oscar_movies=[
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232]
non_oscar_movies=["slumdog Millonaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue=[378,2788,829,185,275]
x=np.arange(len(years))
width=0.4
plt.bar(x-width/2,oscar_revenue,width,label="oscar movies")
plt.bar(x+width/2,non_oscar_revenue,width,label="Non_oscar movies")
plt.title("oscar movies vs non_oscar revenue")
plt.xlabel("Years")
plt.ylabel("Revenue(in SM)")
plt.legend()

plt.xticks(x,years)
```

```python
people=["People A","People B","People C","People D","People E","People F","People G","People H","People I","Person J"]
age=[22,25,30,35,40,45,50,55,60,65]
blood_pressure=[110,115,120,122,115,122,130,135,123,145]
colors=["green" if x<=120 else "red" for x in blood_pressure]
#plt.scatter(age,blood_pressure,color=colors,s=blood_pressure,marker="o",alpha=1)
plt.scatter(age,blood_pressure,s=blood_pressure,cmap="OrRd",c=blood_pressure)
plt.title("age vs BP")
plt.xlabel("Age")
plt.ylabel("Bloood Pressure")
for i in range(len(people)):
    plt.annotate(people[i],xy=(age[i],blood_pressure[i]),xytext=(age[i],blood_pressure[i]))

plt.xlim(min(age)-1,max(age)+7)
```
