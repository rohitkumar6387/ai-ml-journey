```python
df_customers=pd.DataFrame({
    "coustomer_id":[1,2,3,4],
    "name":["bob","adam","farah","dave"]
})

df_orders=pd.DataFrame({
    "order_id":[101,102,103,104],
    "coustomer_id":[2,1,3,5],
    "amount":[250,300,453,125]
})
```

```python
pd.merge(df_customers,df_orders,on="coustomer_id") #innerjoin
```

```python
pd.merge(df_customers,df_orders,on="coustomer_id",how="left") #leftjoin
```

```python
pd.merge(df_customers,df_orders,on="coustomer_id",how="right") #"rightjoin"
```

```python
pd.merge(df_customers,df_orders,on="coustomer_id",how="outer") #outerjoin
```
